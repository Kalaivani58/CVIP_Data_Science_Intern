## ***DATA CLEANING***

```python
need = df[['eventid','iyear','imonth','country_txt','city','suicide','gname']]
need.head()
```

```python
need['gname'].nunique()
```

## ***DATA VISUALIZATION***

```python
plt.figure(figsize = (10,6))
sns.countplot(x = 'iyear',data = need)
plt.xlabel('YEAR')
plt.ylabel('NUMBER OF ATTACKS')
plt.title("TERRORIST ATTACK BY YEAR")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (5,5))
sns.countplot(x = 'imonth',data = need)
plt.xlabel('MONTH')
plt.ylabel('NUMBER OF ATTACKS')
plt.title("TERRORIST ATTACK BY MONTH")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (3,3))
print("0 = NO   1 = YES")
sns.countplot(x = 'suicide',data = need)
plt.xlabel('SUICIDE OR NOT')
plt.ylabel('COUNT')
plt.title("NO. OF SUICIDE")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.countplot(x = 'region_txt',data = df)
plt.xlabel('REGION')
plt.ylabel('NUMBER OF ATTACKS')
plt.title("TERRORIST ATTACK BY REGION")
plt.xticks(rotation=90)
plt.show()
```
