### ***DATA PREPROCESSING***

```python
df.drop('Unnamed: 32',axis = 1, inplace = True)
df.drop('id',axis = 1, inplace = True)
```

```python
df
```

```python
y = df['diagnosis']
x = df.drop('diagnosis',axis = 1)
```

```python
x
```

### ***MODEL TRAINING AND TESTING***

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 24)
```

## *Decision Tree*

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
dt_acc = accuracy_score(y_test,y_pred)
print(dt_acc)
```

## *Support Vector Machine*

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train,y_train)
sy_pred = svm.predict(x_test)
svm_acc = accuracy_score(y_test,sy_pred)
print(svm_acc)
```

## *K Nearest Neighbor Classifier*

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
ky_pred = knn.predict(x_test)
knn_acc = accuracy_score(y_test,ky_pred)
print(knn_acc)
```

## *Naive Bayes Algorithm*

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(x_train,y_train)
ny_pred = nb.predict(x_test)
nb_acc = accuracy_score(y_test,ny_pred)
print(nb_acc)
```

## *Random Forest*

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
ry_pred = rf.predict(x_test)
rf_acc = accuracy_score(y_test,ry_pred)
print(rf_acc)
```

### ***INTERPRETATION***

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (3,3))
alg = ['Decision Tree','SVM','KNN','Naive Bayes','Random Forest']
acc = [dt_acc,svm_acc,knn_acc,nb_acc,rf_acc]
plt.bar(alg, acc,color = 'purple')
plt.xlabel('Algorithms')
plt.ylabel('Accuracy')
plt.title('Different Algorithms and its accuracy score')
plt.xticks(rotation=90)
plt.show()
```
